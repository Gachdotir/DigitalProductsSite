*   Outliers




```python
# Calculate IQR for a specific column (e.g., 'alcohol')
Q1 = red_df['alcohol'].quantile(0.25)
Q3 = red_df['alcohol'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap values outside the bounds

red_pre = red_df.copy()

red_pre['alcohol'] = red_pre['alcohol'].apply(
    lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x)
)
```

```python
# Calculate IQR for 'total sulfur dioxide'
Q1_total = red_pre['total sulfur dioxide'].quantile(0.25)
Q3_total = red_pre['total sulfur dioxide'].quantile(0.75)
IQR_total = Q3_total - Q1_total

lower_bound_total = Q1_total - 1.5 * IQR_total
upper_bound_total = Q3_total + 1.5 * IQR_total

# Calculate IQR for 'free sulfur dioxide'
Q1_free = red_pre['free sulfur dioxide'].quantile(0.25)
Q3_free = red_pre['free sulfur dioxide'].quantile(0.75)
IQR_free = Q3_free - Q1_free

lower_bound_free = Q1_free - 1.5 * IQR_free
upper_bound_free = Q3_free + 1.5 * IQR_free

# Remove outliers for both columns
red_pre = red_pre[(red_pre['total sulfur dioxide'] >= lower_bound_total) & (red_pre['total sulfur dioxide'] <= upper_bound_total)]
red_pre = red_pre[(red_pre['free sulfur dioxide'] >= lower_bound_free) & (red_pre['free sulfur dioxide'] <= upper_bound_free)]
```

```python
red_pre['residual_sugar_log'] = np.log1p(red_pre['residual sugar'])
```

```python
# Before transformation
print(red_pre['residual sugar'].skew())

# After transformation
print(red_pre['residual_sugar_log'].skew())
```
