```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
red_data = pd.read_csv('/content/drive/MyDrive/learning/Python/Machine Learning/Projects/Wine Quality/winequality-red.csv', sep= ';')
white_data = pd.read_csv('/content/drive/MyDrive/learning/Python/Machine Learning/Projects/Wine Quality/winequality-white.csv', sep= ';')
```

```python
red_data.head()
```

```python
white_data.head()
```

```python
red_info = red_data.info()
white_info = white_data.info()
print(red_info)
print(white_info)
```

```python
red_shape = red_data.shape
white_shape = white_data.shape
print(red_shape)
print(white_shape)
```

```python
red_describe = red_data.describe()
white_describe = white_data.describe()
print(red_describe)
print('-----------------------------------------------------------------------')
print(white_describe)
```

```python
red_dup = red_data.duplicated().sum()
white_dup = white_data.duplicated().sum()
print(red_dup)
print(white_dup)
```

```python
red_df = red_data.drop_duplicates()
white_df = white_data.drop_duplicates()
print(red_df.shape)
print(white_df.shape)
```

```python
red_df.hist(bins=20, figsize=(10,10))
plt.tight_layout()
plt.show()
```

```python
white_df.hist(bins=20, figsize=(10,10))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=red_df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=white_df)
plt.xticks(rotation=90)
plt.show()
```
